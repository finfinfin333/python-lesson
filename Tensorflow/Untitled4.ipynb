```python
# coding:utf-8
import tkinter as tk

#円をかくクラス
class Ball: 
    def __init__(self, x, y, dx, dy, color):
        self.x = x
        self.y = y
        self.dx = dx
        self.dy = dy
        self.color = color

    def move(self, canvas): #円を動かすためのメソッド
    #今の円を消す
        canvas.erase(canvas)
    #x,y座標を動かす
        self.x = self.x + self.dx
        self.y = self.y + self.dy
    #次の円を書く
        canvas.draw(canvas)
    #端を超えたら反対向きにする
        if (self.x >= canvas.winfo_width()):
           self.dx = -1
        if (self.x <= 0):
           self.dx = +1
    #y座標も同じ
        if (self.y >= canvas.winfo_height()):
           self.dy = -1
        if (self.y <= 0):
           self.dy = +1

    def erase(self, canvas):
        canvas.erase_oval(self.x - 20, self.y - 20, self.x + 20, self.y + 20, fill = "white", width = 0)
    def draw(self, canvas):
        canvas.draw_oval(self.x - 20, self.y - 20, self.x + 20, self.y + 20, fill = self.color , width = 0)

#四角形を書くクラス
class Rectangle(Ball):
    def erase(self, canvas):
        canvas.erase_rectangle(self.x - 20, self.y - 20, self.x + 20, self.y + 20, fill = "white", width = 0)
    def draw(self, canvas):
        canvas.draw_rectangle(self.x - 20, self.y - 20, self.x + 20, self.y + 20, fill = self.color , width = 0)

#三角形を書くクラス
class Triangle(Ball):
    def erase(self, canvas):
        canvas.erase_polygon(self.x - 20, self.y - 20, self.x + 20, self.y + 20, fill = "white", width = 0)
    def draw(self, canvas):
        canvas.draw_polygon(self.x - 20, self.y - 20, self.x + 20, self.y + 20, fill = self.color , width = 0)

#円、四角形、三角形をまとめて用意
balls = [
    Ball(400, 300, 1, 1, "red"),
    Rectangle(200, 100, -1, 1, "green"),
    Triangle(100, 200, 1, -1, "blue")
]

def loop():
    #動かすと命令
    for b in balls:
        b.move(canvas)
    #もう一回
    root.after(10,loop)

# ウインドウを書く
root = tk.Tk()
root.geometry("600*400")

# Canvas を置く
canvas = tk.Canvas(root, width = 800, height = 600, bg="#fff")
canvas.place(x = 0, y = 0)

# タイマーを設定する
#0.01秒後にmove関数が実行されるようにする
root.after(10, loop)

root.mainloop()
```
